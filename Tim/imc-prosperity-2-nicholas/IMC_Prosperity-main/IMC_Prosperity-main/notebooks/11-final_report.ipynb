```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Final report per round + final portifolio performance

## Round 1 - Pearls & Bananas

```python
data1 = pd.read_csv('../results/results_round1.csv', sep = ';', index_col='timestamp')
```

```python
pnl_pearls = data1.loc[data1['product'] == 'PEARLS']['profit_and_loss']
returns_pearls = pnl_pearls.diff()
sharpe_pearls = returns_pearls.mean() / returns_pearls.std() * np.sqrt(252)

pnl_bananas = data1.loc[data1['product'] == 'BANANAS']['profit_and_loss']
returns_bananas = pnl_bananas.diff()
sharpe_bananas = returns_bananas.mean() / returns_bananas.std() * np.sqrt(252)
```

```python
plt.figure(figsize=(16,8))

plt.subplot(1,2,1)
pnl_pearls.plot(title=f'Profit and loss Pearls - Sharpe {round(sharpe_pearls,2)}', ylabel='Shells')

plt.subplot(1,2,2)
pnl_bananas.plot(title=f'Profit and loss Bananas - Sharpe {round(sharpe_bananas,2)}', ylabel='Shells')
```

## Round 2 - Coconuts and Pina Coladas

```python
data2 = pd.read_csv('../results/results_round2.csv', sep = ';')
```

```python
pnl_coconuts_pina_coladas = data2.loc[data2['product'].isin(['COCONUTS', 'PINA_COLADAS'])][['timestamp', 'profit_and_loss']].groupby('timestamp').sum()['profit_and_loss']
returns_coconuts_pina_coladas = pnl_coconuts_pina_coladas.diff()
sharpe_coconuts_pina_coladas = returns_coconuts_pina_coladas.mean() / returns_coconuts_pina_coladas.std() * np.sqrt(252)
```

```python
plt.figure(figsize=(8,8))
pnl_coconuts_pina_coladas.plot(title=f'Profit and loss Coconuts & Pina Coladas - Sharpe {round(sharpe_coconuts_pina_coladas,2)}', ylabel='Shells')
```

## Round 5 - Final performances 

```python
data5 = pd.read_csv('../results/results_round5.csv', sep = ';')
```

```python
pnl_pearls = data5.loc[data5['product'] == 'PEARLS'].set_index('timestamp')['profit_and_loss']
returns_pearls = pnl_pearls.diff()
sharpe_pearls = returns_pearls.mean() / returns_pearls.std() * np.sqrt(252)

pnl_bananas = data5.loc[data5['product'] == 'BANANAS'].set_index('timestamp')['profit_and_loss']
returns_bananas = pnl_bananas.diff()
sharpe_bananas = returns_bananas.mean() / returns_bananas.std() * np.sqrt(252)

pnl_coconuts_pina_coladas = data5.loc[data5['product'].isin(['COCONUTS', 'PINA_COLADAS'])][['timestamp', 'profit_and_loss']].groupby('timestamp').sum()['profit_and_loss']
returns_coconuts_pina_coladas = pnl_coconuts_pina_coladas.diff()
sharpe_coconuts_pina_coladas = returns_coconuts_pina_coladas.mean() / returns_coconuts_pina_coladas.std() * np.sqrt(252)

pnl_berries = data5.loc[data5['product'] == 'BERRIES'].set_index('timestamp')['profit_and_loss']
returns_berries = pnl_berries.diff()
sharpe_berries = returns_berries.mean() / returns_berries.std() * np.sqrt(252)

pnl_diving_gears = data5.loc[data5['product'] == 'DIVING_GEAR'].set_index('timestamp')['profit_and_loss']
returns_diving_gears = pnl_diving_gears.diff()
sharpe_diving_gears = returns_diving_gears.mean() / returns_diving_gears.std() * np.sqrt(252)

pnl_picnic = data4.loc[data4['product'].isin(['PICNIC_BASKET', 'UKULELE', 'DIP', 'BAGUETTE'])][['timestamp', 'profit_and_loss']].groupby('timestamp').sum()['profit_and_loss']
returns_picnic = pnl_picnic.diff()
sharpe_picnic = returns_picnic.mean() / returns_picnic.std() * np.sqrt(252)
```

```python
plt.figure(figsize=(16,16))

plt.subplot(3,2,1)
pnl_pearls.plot(title=f'Profit and loss Pearls - Sharpe {round(sharpe_pearls,2)}', ylabel='Shells')

plt.subplot(3,2,2)
pnl_pearls.plot(title=f'Profit and loss Banans - Sharpe {round(sharpe_bananas,2)}', ylabel='Shells')

plt.subplot(3,2,3)
pnl_coconuts_pina_coladas.plot(title=f'Profit and loss Coconuts & Pina Coladas - Sharpe {round(sharpe_coconuts_pina_coladas,2)}', ylabel='Shells')

plt.subplot(3,2,4)
pnl_berries.plot(title=f'Profit and loss Berries - Sharpe {round(sharpe_berries,2)}', ylabel='Shells')

plt.subplot(3,2,5)
pnl_diving_gears.plot(title=f'Profit and loss Diving Gear - Sharpe {round(sharpe_diving_gears,2)}', ylabel='Shells')

plt.subplot(3,2,6)
pnl_picnic.plot(title=f'Profit and loss Picnic - Sharpe {round(sharpe_picnic,2)}', ylabel='Shells')
```

```python
pnl_portifolio = data5.loc[~(data5['product'] == 'DOLPHIN_SIGHTINGS')][['timestamp', 'profit_and_loss']].groupby('timestamp').sum()['profit_and_loss']
returns_portifolio = pnl_portifolio.diff()
sharpe_portifolio = returns_portifolio.mean() / returns_portifolio.std() * np.sqrt(252)
```

## Final Portifolio Performance

```python
plt.figure(figsize=(8,8))
pnl_portifolio.plot(title=f'Profit and loss Portifolio - Sharpe {round(sharpe_portifolio,2)}', ylabel='Shells')
```
