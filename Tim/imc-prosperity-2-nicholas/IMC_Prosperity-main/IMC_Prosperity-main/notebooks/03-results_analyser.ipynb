```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename = "/home/nicolas/Documents/projects/IMC_Prosperity/data/results/e7f6a8c4-8433-43ef-aa9f-7d41a333eff4.log"
with open(filename) as file:
    lines = [line.rstrip() for line in file]
```

```python
pnl = []
for idx, line in enumerate(lines):
    if line.startswith("PnL"):
        pnl.append(float(line.split(' ')[1]))
```

```python
position = []
for idx, line in enumerate(lines):
    if line.startswith("Product COCONUTS"):
        position.append(float(line.split(' ')[3].replace(',', '')))
```

```python
pd.Series(position).plot()
```

```python
pd.Series(pnl).plot()
```

```python
pnl_series = pd.Series(pnl)
```

```python
returns = pnl_series.diff()
```
