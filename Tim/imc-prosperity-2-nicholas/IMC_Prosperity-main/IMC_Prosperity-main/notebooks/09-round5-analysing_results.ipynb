```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('../results/results_round1.csv', sep = ';')\
    .set_index("timestamp")
data2 = pd.read_csv('../results/results_round2.csv', sep = ';')\
    .set_index("timestamp")
data3 = pd.read_csv('../results/results_round3.csv', sep = ';')\
    .set_index("timestamp")
data4 = pd.read_csv('../results/results_round4.csv', sep = ';')\
    .set_index("timestamp")
```

```python
PRODUCT = 'PEARLS'

plt.figure(figsize=(24,12))

plt.subplot(2,2,1)
data1.loc[data1['product'] == PRODUCT].profit_and_loss.plot()

plt.subplot(2,2,2)
data2.loc[data2['product'] == PRODUCT].profit_and_loss.plot()

plt.subplot(2,2,3)
data3.loc[data3['product'] == PRODUCT].profit_and_loss.plot()

plt.subplot(2,2,4)
data4.loc[data4['product'] == PRODUCT].profit_and_loss.plot()
```
